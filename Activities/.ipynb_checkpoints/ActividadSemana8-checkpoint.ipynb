# Linear Models


- In supervised learning, the training data fed to the algorithm includes the desired solutions, called labels. 
- In **regression**, the labels are continuous quantities.
- Linear models predict by computing a weighted sum of input features plus a bias term.

```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split

# to make this notebook's output stable across runs
np.random.seed(42)
```

# Ejercicio 1

Utiliza la base de datos de 
https://www.kaggle.com/vinicius150987/manufacturing-cost

Suponga que trabaja como consultor de una empresa de nueva creación que busca desarrollar un modelo para estimar el costo de los bienes vendidos a medida que varían el volumen de producción (número de unidades producidas). La startup recopiló datos y le pidió que desarrollara un modelo para predecir su costo frente a la cantidad de unidades vendidas.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/marypazrf/bdd/main/EconomiesOfScale.csv')
df.sample(10)
```

```python
X = df[['Number of Units']]
y = df['Manufacturing Cost']
```

```python
len(X)
```

```python
y.describe
```

```python
plt.plot(X,y,'b.')
```

### Regression Lineal

Particion enre entrenamiento y test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1) 
```

Diccionario para guardar las métricas

```python
mape_results = {}
r2_results = {}
```

Entrenamiento del modelo

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import r2_score
reg = LinearRegression()
reg.fit(X_train,y_train)
y_train_pred = reg.predict(X_train)
y_test_pred = reg.predict(X_test)
slope = reg.coef_
intercept = reg.intercept_
print('Intercept: ', intercept, ' slope: ', slope)

mape_results['LinearRegression'] = mean_absolute_percentage_error(y_test_pred, y_test)
r2_results['LinearRegression'] = r2_score(y_test_pred, y_test)
```

Modelo de regresion Lineal

$\hat{y} = -6.04X +67.01$

```python
plt.plot(X_train,y_train,'b.')
plt.plot(X_test,y_test,'g.')
plt.plot(X_train, y_train_pred, 'r')
plt.xlabel("X")
plt.ylabel("Y")
```

Podemos ver en la gráfica anterior que la linea no se ajusta muy bien a los datos, dado que la relación parece ser de tipo cuadrático en lugar de lineal

```python
from yellowbrick.regressor import ResidualsPlot
visualizer = ResidualsPlot(reg)
visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
visualizer.show()    
```

En el gráfico anterior podemos ver que el R2, tanto para entrenamiento como para prueba es bastante bajo (alrededor de 58%) y que nuestros errores presentan heterosticidad por lo cual no tenemos un buen ajuste del modelo.

### Regresion Polinomial

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
X_train_poly
```

```python
poly_reg = LinearRegression()
poly_reg.fit(X_train_poly,y_train)
y_train_pred = poly_reg.predict(X_train_poly)
y_test_pred = poly_reg.predict(X_test_poly)
slope = poly_reg.coef_
intercept = poly_reg.intercept_
print('Intercept: ', intercept, ' slope: ', slope)


mape_results['PolyRegression'] = mean_absolute_percentage_error(y_test_pred, y_test)
r2_results['PolyrRegression'] = r2_score(y_test_pred, y_test)
```

Modelo de regresion polinomial

$\hat{y} = 0.054X^4 - 1.36X^3 + 1.27X^2 - 5.55X + 132.15$

```python
plt.plot(X_train,y_train,'b.')
plt.plot(X_test,y_test,'g.')


x_pd = pd.Series(np.array(X_train).flatten())
y_pred_pd = pd.Series(y_train_pred.flatten())

x_sorted = x_pd.sort_values()

Y_pred = np.array(y_pred_pd[x_sorted.index])
x_arr = np.array(x_sorted)

plt.plot(x_arr, Y_pred, color='r')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

En el gráfico anterior podemos ver que la regresion polinomial de grado 4 se ajusta mucho mejor a los datos que la lineal de grado 1.

```python
visualizer = ResidualsPlot(poly_reg)
visualizer.fit(X_train_poly, y_train)  # Fit the training data to the visualizer
visualizer.score(X_test_poly, y_test)  # Evaluate the model on the test data
visualizer.show()   
```

Podemos ver que logramos un mejor R2 al aumentar el grado del polinomio a 4, ahora nuestros errores se comportan de manera mas normal y disminuyó significativamente la heterosticidad, y no hay sobre entrenamiento dado el desempeño se mantiene en los datos de test.

### Ridge Regresion

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=100.0)
ridge.fit(X_train,y_train)
y_train_pred = ridge.predict(X_train)
y_test_pred = ridge.predict(X_test)
slope = ridge.coef_
intercept = ridge.intercept_
print('Intercept: ', intercept, ' slope: ', slope)


mape_results['Ridge'] = mean_absolute_percentage_error(y_test_pred, y_test)
r2_results['Ridge'] = r2_score(y_test_pred, y_test)
```

Modelo de regresion Lineal con Regularizacion L2

$\hat{y} = -5.64X +65.23$

```python
plt.plot(X_train,y_train,'b.')
plt.plot(X_test,y_test,'g.')
plt.plot(X_train, y_train_pred, 'r')
plt.xlabel("X")
plt.ylabel("Y")
```

```python
visualizer = ResidualsPlot(ridge)
visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
visualizer.show()    
```

Al aplicar la regularizacón L2 vemos que el modelo no mejora, esto se debe a que al modelo le falta complejidad, en lugar de sobrarle, por lo cual regularizando no vamos a lograr mejores resultados.

### Lasso Regresion

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=1)
lasso.fit(X_train,y_train)
y_train_pred = lasso.predict(X_train)
y_test_pred = lasso.predict(X_test)
slope = lasso.coef_
intercept = lasso.intercept_
print('Intercept: ', intercept, ' slope: ', slope)

mape_results['Lasso'] = mean_absolute_percentage_error(y_test_pred, y_test)
r2_results['Lasso'] = r2_score(y_test_pred, y_test)
```

$\hat{y} = -5.48X +64.48$

```python
plt.plot(X_train,y_train,'b.')
plt.plot(X_test,y_test,'g.')
plt.plot(X_train, y_train_pred, 'r')
```

```python
visualizer = ResidualsPlot(lasso)
visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
visualizer.show()    
```

Al aplicar la regularizacón L1 vemos que el modelo no mejora, esto se debe a que al modelo le falta complejidad, en lugar de sobrarle, por lo cual regularizando no vamos a lograr mejores resultados.

### Comparacion Modelos

```python
plt.bar(*zip(*mape_results.items()))
plt.title('MAPE')
plt.show()
```

```python
plt.bar(*zip(*r2_results.items()))
plt.title('R2')
plt.show()
```

El modelo que más le serviría a la empresa entre estos 4, sería la regresión lineal polinomial, dado que muestra mejor desempeño en ambas métricas (MAPE 11%, R2 60%), tanto en entrenamiento como en test, y no muestra signos de sobreentrenamiento. Sin embargo, aun se tendría un % de variabilidad no explicada del 40% lo cual es bastante alto y nos estaría indicando que a lo mejor faltan variables adicionales que nos permitan predecir adecuadamente el valor de producir una unidad.

### Clustering con Kmeans

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
kmeans = KMeans( n_init=20)

visualizer = KElbowVisualizer(kmeans, k=(2,30),metric='distortion',timings=False,)
visualizer.fit(gdf[['latitude','longitude']])  
visualizer.show() 
```

Del gráfico anterior vemos que una buena cantidad de almaecenes a poner sería 8, dado que después de esta cifra se reduce considerablemente en el impacto en la disminución de recorridos por cada nuevo almacén. Sin embargo, está es solo una aproximación inicial, para determinar el número optimo sería necesario conocer a profunidad los costos de transporte y los costos de construir y operar cada nuevo almacén.

Este cálculo se hizo mediante la métrica Distorcion que calcula la suma de las distancias de cada tienda a su respectivo almacén al cuadrado, dado que la suma de las distancias es lo que nos gustaría minimizar para reducir los costos de transporte de la compañía.

```python
kmeans = KMeans(n_init=20,n_clusters=8)
y_pred = kmeans.fit_predict(gdf[['latitude','longitude']])
```

```python
almacenes = kmeans.cluster_centers_
almacenes = pd.DataFrame(almacenes , columns=['latitude', 'longitude'])
almacenes["Coordinates"] = list(zip(almacenes.longitude, almacenes.latitude))
almacenes["Coordinates"] = almacenes["Coordinates"].apply(Point)
almacenes_geo = gpd.GeoDataFrame(almacenes, geometry="Coordinates")
almacenes_geo
```

```python

fig, gax = plt.subplots(figsize=(10,10))

world.query("name == 'United States of America'").plot(ax = gax, edgecolor='black', color='white')


gdf.plot(ax=gax, color='red', alpha = 0.5)
almacenes_geo.plot(ax=gax, color='blue')

gax.set_xlabel('longitude')
gax.set_ylabel('latitude')
gax.set_title('Target en Estados Unidos')

gax.spines['top'].set_visible(False)
gax.spines['right'].set_visible(False)

plt.show()
```

En el mapa anterior podemos ver la ubicación de los  almacenes en Estados Unidos. Como vemos, el modelo de Clustering parece haber obtenido buenos resultados, dado que los almacenes quedaron abarcando espacios geográficos diferentes. En Alaska no quedaría ningún almacen debido a los pocos puntos que hay alrededor pero sería necesario evaluar desde los costos mencionados anteriormente si vale la pena o no ubicar allí un almacén.

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geoapiExercises")

def city_state(row):
    coord = f"{row['latitude']}, {row['longitude']}"
    location = geolocator.reverse(coord)
    address = location.raw['address']
    state = address.get('state', '')
    row['state'] = state
    return row


almacenes_geo= almacenes_geo.apply(city_state, axis=1)
almacenes_geo.loc[5,'state']= 'Hawaii'
almacenes_geo['id_almacen']= almacenes_geo.index
almacenes_geo
```

```python
almacenes_geo.columns = ['latitude_almacen', 'longitude_almacen', 'Coordinates_almacen', 'state_almacen', 'id_almacen']
gdf['id_almacen'] = y_pred
gdf = gdf.merge(almacenes_geo,how='left',on = 'id_almacen')
gdf.head()
```

```python
sum_almacenes = gdf.groupby(['state_almacen','latitude_almacen','longitude_almacen']).size()
sum_almacenes
```

En la tabla anterior podemos ver las ubicaciones y el  numero de tiendas que atendería cada almacén.
